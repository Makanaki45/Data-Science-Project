```python
employee.departmentName.unique() #WHAT ARE THE VARIOUS DEPARTMENTS WE HAVE IN THE ORGANIZATION
```

```python
employee.departmentName.value_counts() #HOW MANY EMPLOYEES DO WE HAVE IN EACH DEPARTMENT?
```

```python
ax = employee.departmentName.value_counts().plot(kind='bar', figsize=(10,6))
ax.set_title("Number of Employees in each Department")
ax.set_ylabel("Number of Employees")
ax.set_xlabel("Departments")
```

```python
employee.qualification.unique() #WHAT ARE THE UNIQUE QUALIFICATIONS OF EMPLOYEES THAT WE HAVE
```

```python
employee.qualification.value_counts() #HOW MANY EMPLOYEES DO WE HAVE IN EACH QUALIFICATION CATEGORY?
```

```python
ax = employee.qualification.value_counts().plot(kind='bar', figsize=(10,6))
ax.set_title("Number of Employees in each Qualification")
ax.set_ylabel("Number of Employees")
ax.set_xlabel("Qualification")
```

```python
employee.nationality.unique() #WHAT ARE THE UNIQUE COUNTRIES THAT OUR EMPLOYEES ARE FROM
```

```python
employee.nationality.value_counts() #HOW MANY EMPLOYEES DO WE HAVE IN EACH COUNTRY
```

```python
ax = employee.nationality.value_counts().plot(kind='bar', figsize=(10,6))
ax.set_title("Number of Employees From each country")
ax.set_ylabel("Number of Employees")
ax.set_xlabel("Nationality")
```
